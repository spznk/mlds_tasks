# HW7. Numpy

### 1. Створення масивів

```python
import numpy as np
```

```python
print(np.array([5, 15, 25, 35, 45]))
print(np.zeros(7))
print(np.ones(6))
print(np.arange(10, 20))
```

### 2. Базові операції

```python
my_array = np.array([12, 24, 36, 48, 60])
display(my_array)
```

```python
print(my_array * 2)
print(my_array + 10)
print(my_array ** 2)
print(np.sqrt(my_array))
```

### 3. Статистичні функції

```python
data = np.array([18, 27, 33, 41, 52, 68, 79])
display(data)
```

```python
print(data.mean())
print(data.min())
print(data.max())
print(data.sum())
```

### 4. Індексація

```python
arr = np.array([15, 22, 35, 44, 58, 67])
display(arr)
```

```python
print(arr[0])
print(arr[-1])
print(arr[1:4])

mask = arr > 40
print(arr[mask])
```

### 5. Двовимірні масиви

```python
matrix = np.array([[2, 4, 6],
                   [8, 10, 12],
                   [14, 16, 18]])
display(matrix)
```

```python
print(matrix.shape)
print(matrix[0, :])
print(matrix[:, 1])
print(matrix.sum())
print(matrix.mean())
```
