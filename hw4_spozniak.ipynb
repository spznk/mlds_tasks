### 1. Цикл for - базовий

```python
for i in range(1, 8):
    print(i)
```

### 2. Цикл for по списку

```python
names = ["Олег", "Анна", "Іван", "Марія"]
```

```python
for name in names:
    print(f"Привіт, {name}!")
```

### 3. Цикл з enumerate

```python
grades = [85, 92, 78, 95, 88]
```

```python
for i, grade in enumerate(grades):
    print(f"Студент {i + 1}: {grade} балів")
```

### 4. Сума парних чисел

```python
even_sum = 0
for i in range(1, 21):
    if i % 2 == 0:
        even_sum += i

print (even_sum)
```
