### 1. Основи роботи з умовними конструкціями

```python
cooking_time = 5
```

```python
# Просто так цікавіше :)

import random

cooking_time = random.randrange(0, 12)
print(cooking_time)
```

```python
if cooking_time < 4:
    print("Яйце рідке")
elif cooking_time < 7:
    print("Яйце всмятку")
elif cooking_time < 10:
    print("Яйце вкруту")
else:
    print("Яйце переварене")
```

### 2. Перевірка оцінки

```python
grade = 72
```

```python
grade = random.randrange(0, 100)
print(grade)
```

```python
if grade >= 90:
    print("Відмінно")
elif grade >= 75:
    print("Добре")
elif grade >= 60:
    print("Задовільно")
else:
    print("Незадовільно")
```
