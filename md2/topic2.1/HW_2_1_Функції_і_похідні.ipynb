1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
import sympy.plotting as sp_plt
```

```python
# Calculate derivative with Sympy and who result
x = sp.symbols('x')
f = 10*x - 5
df = sp.diff(f, x)

display("Функція:", f, "Похідна:", df)
```

```python
# Convert into Numpy values and build plot
x = np.linspace(0, 10, 11)
f = 10*x - 5
df = x*0 + 10

plt.figure()
plt.plot(x, f, label="f(x) = 10x - 5", color="green")
plt.plot(x, df, label="f'(x) = 10", color="purple")

plt.xlabel("hours learned")
plt.ylabel("score")
plt.legend()
plt.grid(True)
plt.show()
```

**Спостереження**
- Оцінка росте лінійно зі зростом кількості годин, витрачених на навчання

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
# Calculate derivatives with Sympy
x = sp.symbols('x')

f_sin = sp.sin(x)
f_ln = sp.ln(x)

df_sin = sp.diff(f_sin, x)
df_ln = sp.diff(f_ln, x)

# Show results
display("Функція:", f_sin, "Похідна:", df_sin)
print("\n")
display("Функція:", f_ln, "Похідна:", df_ln)
```

```python
# Convert into Numpy values and build plot for f(x) = sin(x)
x = np.linspace(0, 10, 100)
f_sin = np.sin(x)
df_sin = np.cos(x)

plt.figure(figsize=(10, 5))
plt.plot(x, f_sin, label="f(x) = sin(x)", color="blue")
plt.plot(x, df_sin, label="f'(x) = cos(x)", color="green")
plt.legend()
plt.show()
```

```python
# Convert into Numpy values and build plot for f(x) = ln(x)
x = np.linspace(0.1, 20, 100)
f_ln = np.log10(x)
df_ln = 1/x

plt.figure(figsize=(10, 5))
plt.plot(x, f_ln, label="f(x) = ln(x)", color="blue")
plt.plot(x, df_ln, label="f'(x) = 1/x", color="green")
plt.legend()
plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
# Calculate derivative with Sympy and show it
x = sp.symbols('x')
f = x**2 - 5*x + 1
df = sp.diff(f, x)

display("Функція:", f, "Похідна:", df)
```

```python
# Display with sympy.plotting
fig = sp_plt.plot(f, df, (x, 0, 100), legend=True, size=(5, 5), show=False)
fig[0].label = "f"
fig[1].label = "df"
fig.show()
```
