### 1. Цикл while

```python
counter = -3
```

```python
while counter <= 5:
    print(counter)
    counter += 1
```

### 2. List comprehension - прості операції

```python
numbers = [11, 4, 9, 16, 7, 5]

squares = [x**2 for x in numbers]
even_nums = [x for x in numbers if x % 2 == 0]
tenfold_nums = [x * 10 for x in numbers]

print(f"Numbers: {numbers}")
print(f"Number squares: {squares}")
print(f"Even numbers: {even_nums}")
print(f"Numbers multiplied by 10: {tenfold_nums}")
```

### 3. List comprehension - комбінована логіка

```python
numbers = list(range(1, 11))

numbers_new = [x * 2 if x % 2 == 0 else x - 1 for x in numbers]
print(numbers_new)
```
